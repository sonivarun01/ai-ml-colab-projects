```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
```

```python
data = {
    "Age" : [19,34,12,22,31,44],
    "Income": [20000, 30000, 40000, 50000, 60000,34000]
}
```

```python
df = pd.DataFrame(data)
```

```python
print("Original DataFrame:")
print(df)
```

```python
scaler = StandardScaler()
df['Age_standardized'] = scaler.fit_transform(df[['Age']])
df['Income_standardized'] = scaler.fit_transform(df[['Income']])
```

```python
print("\nDataFrame after standardization:")
print(df)
```

```python
print("\nMean and Standard Deviation after standardization:")
print(f"Age mean: {df['Age_standardized'].mean():.2f}, std: {df['Age_standardized'].std():.2f}")
print(f"Income mean: {df['Income_standardized'].mean():.2f}, std: {df['Income_standardized'].std():.2f}")
```
