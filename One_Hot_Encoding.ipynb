```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
```

```python
data = {
    'City': ['New York', 'Los Angeles', 'Chicago', 'Houston', 'Phoenix'],
    'Purchased': ['Yes', 'No', 'No', 'Yes', 'Yes']
}
```

```python
df = pd.DataFrame(data)
```

```python
print("Original DataFrame:")
print(df)
```

```python
onehot_encoder = OneHotEncoder(sparse_output=False)
```

```python
city_encoded = onehot_encoder.fit_transform(df[['City']])
```

```python
city_encoded_df = pd.DataFrame(city_encoded, columns=onehot_encoder.get_feature_names_out(['City']))
```

```python
df_sklearn_encoded = pd.concat([df, city_encoded_df], axis=1)
```

```python
print("\nDataFrame After One-Hot Encoding Using sklearn:")
print(df_sklearn_encoded)
```
